```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import preprocessing
import seaborn as sns
```

```python
runX = pd.read_csv("../data/runX_quant.txt",sep = '\t')
```

```python
runX.head()
```

```python
total_numreads = runX['NumReads'].sum()
```

```python
runX['FPKM'] = [n * 10**3 * 10**6/(total_numreads*l) for (n,l) in zip(runX.NumReads, runX.EffectiveLength)]
```

```python
runX['normalized_reads']= preprocessing.normalize([np.array(runX['NumReads'])])[0]
```

```python
sns.pairplot(runX[['EffectiveLength','TPM','FPKM','normalized_reads']])
plt.savefig("part")
```
